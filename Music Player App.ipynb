```python
from pygame import mixer
from tkinter import *
from tkinter import filedialog
from tkinter import ttk
import os

# Initialize mixer 
mixer.init()

# Define functions
def addsongs():
    temp_songs = filedialog.askopenfilenames(initialdir=".", title="Choose a song", filetypes=(("mp3 Files", "*.mp3"),))
    for song_path in temp_songs:
        song_name = os.path.basename(song_path)
        destination = os.path.join("Music", song_name)
        if not os.path.exists(destination):  # Check if the file doesn't already exist in the Music folder
            shutil.copy(song_path, destination)  # Copy the selected file to the Music folder
            songs_list.insert(END, song_name)  # Add the song name to the listbox

def deletesong():
    curr_song = songs_list.curselection()
    songs_list.delete(curr_song[0])

def Play():
    song = songs_list.get(ACTIVE)
    song_path = os.path.abspath(os.path.join("Music", song))
    mixer.music.load(song_path)
    mixer.music.play()

def Pause():
    mixer.music.pause()

def Stop():
    mixer.music.stop()
    songs_list.selection_clear(ACTIVE)

def Resume():
    mixer.music.unpause()

def Previous():
    previous_one = songs_list.curselection()
    previous_one = previous_one[0] - 1
    temp2 = songs_list.get(previous_one)
    temp2 = f'Music/{temp2}'
    mixer.music.load(temp2)
    mixer.music.play()
    songs_list.selection_clear(0, END)
    songs_list.activate(previous_one)
    songs_list.selection_set(previous_one)

def Next():
    next_one = songs_list.curselection()
    next_one = next_one[0] + 1
    temp = songs_list.get(next_one)
    temp = f'Music/{temp}'
    mixer.music.load(temp)
    mixer.music.play()
    songs_list.selection_clear(0, END)
    songs_list.activate(next_one)
    songs_list.selection_set(next_one)

def set_vol(val):
    volume = float(val) / 100
    mixer.music.set_volume(volume)

# Create the root window 
root = Tk()
root.title('Melody Music Player-by Nishit Shivdasani')

# Set the window size and position
window_width = 900
window_height = 400
screen_width = root.winfo_screenwidth()
screen_height = root.winfo_screenheight()
x_coordinate = (screen_width/2) - (window_width/2)
y_coordinate = (screen_height/2) - (window_height/2)
root.geometry('%dx%d+%d+%d' % (window_width, window_height, x_coordinate, y_coordinate))

# Create a style for the buttons
button_style = ttk.Style()
button_style.configure('TButton', font=('Helvetica', 12), padding=5)

# Create a style for the labels
label_style = ttk.Style()
label_style.configure('TLabel', font=('Helvetica', 12))

# Create the listbox to contain songs
songs_list = Listbox(root, selectmode=SINGLE, bg="black", fg="white", font=('Arial', 12), height=12, width=47, selectbackground="gray", selectforeground="black")
songs_list.grid(columnspan=9, padx=10, pady=10)

# Add all songs already stored in the "Music" folder to the listbox
music_files = os.listdir("Music")
for song_file in music_files:
    songs_list.insert(END, song_file)

# Create buttons
play_button = ttk.Button(root, text="Play", command=Play)
play_button.grid(row=1, column=0, padx=5, pady=5)

pause_button = ttk.Button(root, text="Pause", command=Pause)
pause_button.grid(row=1, column=1, padx=5, pady=5)

stop_button = ttk.Button(root, text="Stop", command=Stop)
stop_button.grid(row=1, column=2, padx=5, pady=5)

resume_button = ttk.Button(root, text="Resume", command=Resume)
resume_button.grid(row=1, column=3, padx=5, pady=5)

previous_button = ttk.Button(root, text="Previous", command=Previous)
previous_button.grid(row=1, column=4, padx=5, pady=5)

next_button = ttk.Button(root, text="Next", command=Next)
next_button.grid(row=1, column=5, padx=5, pady=5)

# Create volume slider
volume_slider = ttk.Scale(root, from_=0, to=100, orient=HORIZONTAL, command=set_vol)
volume_slider.set(70)  # set default volume
volume_slider.grid(row=1, column=6, padx=5, pady=5)

# Create menu
my_menu = Menu(root)
root.config(menu=my_menu)
add_song_menu = Menu(my_menu)
my_menu.add_cascade(label="Menu", menu=add_song_menu)
add_song_menu.add_command(label="Add songs", command=addsongs)
add_song_menu.add_command(label="Delete song", command=deletesong)

# Main loop
root.mainloop()
```
